# SUBPLOT

```python
import matplotlib.pyplot as plt
```

```python
x = [2, 4, 6, 8, 10]
y1 = [i*2 for i in x]
y2 = [i**2 for i in x]

plt.subplot(1,2,1)
plt.plot(x,y1)
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y1)

plt.subplot(1, 2, 2)
plt.plot(x, y2)
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.xlabel("Numbers")
plt.ylabel("x*2")
plt.title("Double Number")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.xlabel("Numbers")
plt.ylabel("x**2")
plt.title("Squared Number")

plt.tight_layout()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.xlabel("Numbers")
plt.ylabel("x*2")
plt.title("Double Number")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.xlabel("Numbers")
plt.ylabel("x**2")
plt.title("Squared Number")

plt.subplot(2, 2, 3)
plt.plot(x, y1)
plt.xlabel("Numbers")
plt.ylabel("x*2")
plt.title("Double Number")

plt.subplot(2, 2, 4)
plt.plot(x, y2)
plt.xlabel("Numbers")
plt.ylabel("x**2")
plt.title("Squared Number")

plt.tight_layout()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 5))

axs[0, 0].plot(x, y1)
axs[0, 0].set_title(" Double Number")

axs[0, 1].plot(x, y2)
axs[0, 1].set_title(" Sqaure Number")

axs[1, 0].plot(x, y1)
axs[1, 0].set_title(" Double Number")

axs[1, 1].plot(x, y2)
axs[1, 1].set_title(" Square Number")

fig.suptitle("Subplot", fontsize=15)
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].plot(x, y1)
axs[0, 0].set_title(" Double Number")

axs[0, 1].plot(x, y2)
axs[0, 1].set_title(" Sqaure Number")

axs[1, 0].plot(x, y1)
axs[1, 0].set_title(" Double Number")

axs[1, 1].plot(x, y2)
axs[1, 1].set_title(" Square Number")

fig.subplots_adjust(top=0.9)
fig.suptitle("Subplot", fontsize=15)
plt.show()
plt.tight_layout()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
y3 = [i**3 for i in x]
y4 = [i**0.5 for i in x]
y = [y1, y2, y3, y4]
title= ["Double", "Square", "Cube", "Square root"]

for i in range(2):
    for j in range(2):
        indx = i*2+j
        axs[i][j].plot(x, y[indx])
        axs[i][j].set_title(title[indx])
        
```
