# SCATTER PLOT

```python
import matplotlib.pyplot as plt

study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
```

```python
plt.scatter(study_hours, exam_scores)
```

```python
plt.scatter(study_hours, exam_scores)
plt.xlabel("study_hours")
plt.ylabel("exam_scores")
plt.title("Study Hours VS Exam Score")
plt.grid(True)
```

```python
sizes = [ score*2 for score in exam_scores]

plt.scatter(study_hours, exam_scores, s=sizes)
plt.xlabel("study_hours")
plt.ylabel("exam_scores")
plt.title("Study Hours VS Exam Score")
plt.grid(True)
```

```python
colors = [ "red" if score>65 else "green" for score in exam_scores] 

plt.scatter(study_hours, exam_scores, s=sizes, c=colors)
plt.xlabel("study_hours")
plt.ylabel("exam_scores")
plt.title("Study Hours VS Exam Score")
plt.grid(True)
```

```python
plt.scatter(study_hours, exam_scores, c=exam_scores, cmap="spring_r")
plt.xlabel("study_hours")
plt.ylabel("exam_scores")
plt.title("Study Hours VS Exam Score")
plt.grid(True)
```

Adding Annotations

```python
for i in range(len(study_hours)):
    plt.annotate(f'S{i+1}',(study_hours[i]+0.1,exam_scores[i]-0.3))

plt.scatter(study_hours, exam_scores, c=exam_scores, cmap="spring_r")
plt.xlabel("study_hours")
plt.ylabel("exam_scores")
plt.title("Study Hours VS Exam Score")
plt.grid(True)
```

```python
class_a_hours = [1, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]

class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]

plt.scatter(class_a_hours, class_a_scores, label='Class A', color='red')
plt.scatter(class_b_hours, class_b_scores, label='Class B', color='blue')

plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
