```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs)
```

```python
plt.bar(years, runs)
plt.title("ST Runs")
plt.xlabel("Runs")
plt.ylabel("Years")
plt.tight_layout()
```

```python
import numpy as np

sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
sehwag = [0, 200, 900, 1400, 1600, 1800, 1500, 1100, 800, 0]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]

x = np.arange(len(years))
width = 0.25

plt.bar(x - width, sachin, width=0.25, label="sachin")
plt.bar(x, sehwag, width=0.25, label="sehwag")
plt.bar(x + width, kohli, width=0.25, label="kolhi")
plt.title("ST Runs")
plt.xlabel("Runs")
plt.ylabel("Years")
plt.legend()
plt.tight_layout()
```

```python
plt.bar(x - width, sachin, width=0.25, label="sachin")
plt.bar(x, sehwag, width=0.25, label="sehwag")
plt.bar(x + width, kohli, width=0.25, label="kolhi")
plt.title("ST Runs")
plt.xlabel("Runs")
plt.ylabel("Years")
plt.legend()
plt.tight_layout()

plt.xticks(x,years)
```

```python
name = ["sachin","sehwag","virat"]
runs = [19000, 10000, 15000]

plt.barh(name,runs, color="skyblue")
plt.xlabel("Years")
plt.title("Totals Runs in Span of 5years")
plt.tight_layout()
```

```python
name = ["sachin","sehwag","virat"]
runs = [19000, 13000, 15000]

for i in range(len(runs)):
    plt.text(i, runs[i] + 30, str(runs[i]), ha="center")
    

plt.bar(name,runs, color="skyblue")
plt.xlabel("Years")
plt.title("Totals Runs in Span of 5years")
plt.tight_layout()
```
