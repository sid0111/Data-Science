```python
import pandas as pd
```

```python
df = pd.read_csv('mobile.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop('img', axis=1, inplace=True)
```

```python
df['3G'] = df['sim'].str.contains('3G', case=False).astype(int)
df['4G'] = df['sim'].str.contains('4G', case=False).astype(int)
df['5G'] = df['sim'].str.contains('5G', case=False).astype(int)
df['VoLTE'] = df['sim'].str.contains('VoLTE', case=False).astype(int)
df['Vo5G'] = df['sim'].str.contains('Vo5G', case=False).astype(int)
df['Wi-Fi'] = df['sim'].str.contains('Wi-Fi', case=False).astype(int)
df['NFC'] = df['sim'].str.contains('NFC', case=False).astype(int)
```

```python
df
```

```python
df.head(20)
```

```python
df.head(10)
```

```python
df.drop('sim', axis=1, inplace=True)
```

```python
df
```

```python
df.dropna(subset=['storage'], inplace=True)
```

```python
df
```

```python
df['Processor_Name'] = df['processor'].str.split(',').str[0]
```

```python
df
```

```python
df['Clock_Speed_GHz'] = df['processor'].str.split(',').str[3]
```

```python
df
```

```python
df.drop('Clock_Speed_GHz', inplace=True, axis=1) 
```

```python
df
```

```python
df['Clock_Speed_GHz'] = df['processor'].str.extract(r'([\d.]+)\s*GHz').astype(float)
```

```python
df
```

```python
df.drop('processor', inplace=True, axis=1)
```

```python
df
```

```python
df[['Ram_GB','Storage_GB']] = df['storage'].str.extract(r'(\d+)\s*GB RAM.*?(\d+)\s*GB')
```

```python
df
```

```python
df.drop('RAM',inplace=True, axis = 1)
```

```python
df
```

```python
df['Battery_mAh'] = df['battery'].str.extract(r'(\d+)\s*mAh')
df['Charging_Watt'] = df['battery'].str.extract(r'(\d+)\s*W')
df['Fast_charging'] = df['battery'].str.contains('Fast', case=False)
```

```python
df
```

```python
df.drop(['storage','battery'], inplace=True, axis=1)
```

```python
df
```

```python
df.drop(['memoryExternal','fm'], inplace=True, axis=1)
```

```python
df
```

```python
df['Display_Size_inch'] = df['display'].str.extract(r'([\d.]+)\s*inches').astype(float)
```

```python
df
```

```python
df['Refresh_Rate_Hz'] = df['display'].str.extract(r'(\d+)\s*Hz')
df
```

```python
df['Back Camera'] = df['camera'].str.extract(r'(\d+)\s*MP')
```

```python
df
```

```python
df['Front_Camera'] = df['camera'].str.extract(r'(\d+)\s*MP Front Camera')
```

```python
df
```

```python
df.drop(['display'], axis=1, inplace=True)
```

```python
df
```

```python
df.to_csv('cleaned_mobile_dataset.csv', index=False)
```
