# Performance Score
- RAM × Clock Speed × Processor Tier

```python
pd.reset_option('display.max_rows')
```

```python
tier_3 = [
    "Dimensity 9200 Plus",
    "Snapdragon 8 Elite",
    "Snapdragon 8 Gen3",
    "Snapdragon 8 Gen1",
    "Dimensity 9300 Plus",
    "Snapdragon 888",
    "Snapdragon 870",
    "Snapdragon 855+",
    "Dimensity 1000+"
]
```

```python
tier_2 = [
    "Snapdragon 778G",
    "Dimensity 8300",
    "Snapdragon 6 Gen4",
    "Helio G99",
    "Dimensity 7050",
    "Exynos 1380",
    "Snapdragon 695",
    "Snapdragon 7 Gen3",
    "Dimensity 7300"
]
```

```python
tier_1 = [
    "Dimensity 6300",
    "Octa Core Processor",
    "Dimensity 6020",
    "Helio P35",
    "Unisoc T612",
    "Snapdragon 4 Gen1",
    "Helio G85",
    "Unisoc T606",
    "Exynos 850"
]
```

```python
def get_proc_tier(name):
    name = str(name)
    if any(key in name for key in tier_1):
        return 3
    elif any(key in name for key in tier_2):
        return 2
    elif any(key in name for key in tier_3):
        return 1
    else:
        return 1
```

```python
df['Processor_Tier'] = df['Processor_Name'].apply(get_proc_tier)
```

```python
df
```

```python
df.head(100)
```

```python
df['Performance_Score'] = df['Ram_GB'] * df['Clock_Speed_GHz'] * df['Processor_Tier']
```

```python
df
```

# Battery Efficiency

```python
df['Battery_Efficiency'] = df['Battery_mAh'] / df['Display_Size_inch']
```

```python
df
```

```python
connectivity_features = ['3G', '4G', '5G', 'VoLTE', 'Vo5G', 'Wi-Fi', 'NFC']
df['Connectivity_Score'] = df[connectivity_features].sum(axis=1)
```
