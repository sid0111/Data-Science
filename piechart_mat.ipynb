# PIECHART

```python
sid = [8, 7, 9, 8, 6]
day = ['mon', 'tues', 'wed', 'thurs', 'frid']

plt.pie(sid, labels=day)
```

```python
sid = [8, 7, 9, 8, 6]
day = ['mon', 'tues', 'wed', 'thurs', 'frid']

explode = [0.05, 0.05, 0.2, 0.05, 0.05]

plt.pie(sid, labels=day, explode=explode)
plt.title("No. of Hours Studying")
```

```python
plt.pie(sid, labels=day, explode=explode, autopct='%1.1f%%')
plt.title("No. of Hours Studying")
```

```python
plt.pie(sid, labels=day, explode=explode, autopct='%d%%')
plt.title("No. of Hours Studying")
```

```python
plt.pie(sid, labels=day, explode=explode, autopct='%d%%', startangle=180)
plt.title("No. of Hours Studying")
```

```python
plt.pie(
    sid, 
    labels=day, 
    explode=explode, 
    autopct='%d%%', 
    startangle=180,
    wedgeprops ={'edgecolor':'black', 'linewidth':2, 'linestyle':'--'}
)
plt.title("No. of Hours Studying")
```

```python
plt.pie(
    sid, 
    labels=day, 
    explode=explode, 
    autopct='%d%%', 
    #startangle=180,
    #wedgeprops ={'edgecolor':'black', 'linewidth':2, 'linestyle':'--'},
    shadow=True
)
plt.title("No. of Hours Studying")
```

# STACKPLOTS

```python
days = [1, 2, 3, 4, 5]
studying = [8, 7, 7, 6, 8]
sleeping = [7, 6, 8, 6, 8]
playing = [2, 3, 2, 1, 4]
gym = [1, 2, 1, 2, 1]

labels = ['studying', 'sleeping', 'playing', 'gym']

plt.stackplot(days, studying, sleeping, playing, gym, labels=labels)
```

```python
plt.figure(figsize=(10,10))
plt.stackplot(days, studying, sleeping, playing, gym, labels=labels)

plt.legend(loc='upper left')
plt.title("Hours distribution")
plt.xlabel("Day")
plt.ylabel("Hours")
```
